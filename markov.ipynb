**Modifier le modèle**

```python
#Q1
pi_0=np.array([1,0,0])
matrice=np.array([[0.08,0.92,0],[0,0.93,0.07],[0.02,0,0.98]])
nb_jours=150
nb_individus=200
s=projet.seq_n(nb_jours,matrice,pi_0,nb_individus)
(liste_t,liste_s,liste_i,liste_r)=projet.nb_indiv_s_i_r(s,nb_jours)
print("La nouvelle matrice :")
print(matrice)
print("Distribution observer sur une population de 200 individus")
projet.graphe_evolution(liste_t,liste_s,liste_i,liste_r)
print("Pourcentag d'une distribution observer sur une population de 200 individus")
projet.graphe_pourcentage_pratique(liste_t,liste_s,liste_i,liste_r)
print("Repartition théorique des effectifs")
projet.graphe_proba(pi_0,matrice,nb_jours)
print("Distribution théorique et la distribution observée de la longueur de l'immunité")
projet.graphe_distribution_r(matrice[2][0],s)
print("la distribution stationnaire est :")
print(projet.distributionStationnaire(pi_0,matrice))

```

On remarque que plus on augmente la probabilité de sain à infecté, plus les individus deviennent infectés plus rapidement.
On remarque aussi que la longueur de l'imunité observé se rapproche de la théorique même pour un temps inférieur à 20 (ce qui est normal vue que on ce retrouve plus vite à l'état R)

**Confinement**

```python
pi_0=np.array([1,0,0])
matrice_dec=np.array([[0.92,0.08,0],[0,0.93,0.07],[0.02,0,0.98]])
matrice_conf=np.array([[1,0,0],[0,0.93,0.07],[0.02,0,0.98]])
nb_jours=150
nb_indiv=200
print("Matrice de transitions confinement")
print(matrice_conf)
print("Matrice de transitions déconfinement")
print(matrice_dec)

```

```python
s,nb_conf,nb_deconf,temps_conf,temps_dec=projet.gener_seq(pi_0,matrice_dec,matrice_conf,nb_indiv,nb_jours)
(liste_t,liste_s,liste_i,liste_r)=projet.nb_indiv_s_i_r(s,nb_jours)
projet.graphe_evolution(liste_t,liste_s,liste_i,liste_r)
for i in range(len(temps_conf["debut"])):
  print("Confinement "+str(i+1))
  if i < len(temps_conf["fin"]):
    print("\t debut : "+str(temps_conf["debut"][i])+ ", fin : "+str(temps_conf["fin"][i]))
  else :
    print("\t debut : "+str(temps_conf["debut"][i])+ ", fin : en cours")
print()

for i in range(len(temps_dec["debut"])):
  print("Deconfinement "+str(i+1))
  if i < len(temps_dec["fin"]):
    print("\t debut : "+str(temps_dec["debut"][i])+ ", fin : "+str(temps_dec["fin"][i]))
  else :
    print("\t debut : "+str(temps_dec["debut"][i])+ ", fin : en cours")
```
